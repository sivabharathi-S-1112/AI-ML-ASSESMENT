```python
import pandas as pd
```

```python
#sample dataset
data={
    'customerID':[1,2,3,4,5,6,7,8],
    'AnnualIncome':[15,16,17,28,30,45,55,60],
    'SpendingScore':[39,81,6,77,40,50,42,30]
}
```

```python
df=pd.DataFrame(data)
df
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#select features for clustering
x=df[['AnnualIncome','SpendingScore']]

#Initial Scatter plot
plt.figure(figsize=(6,4))
plt.scatter(x['AnnualIncome'], x['SpendingScore'],s=100,c='#90EE90',edgecolors='k')
plt.title('Customer Distribution')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.grid(True)
plt.show()
```

```python
#apply K-Means Clustering
kmeans=KMeans(n_clusters=3,random_state=0)
df['Cluster']=kmeans.fit_predict(x)

#Clustered data Plot
plt.figure(figsize=(6,4))
sns.scatterplot(data=df,x='AnnualIncome',y='SpendingScore',hue='Cluster',palette='Set1',s=100)
centroids=kmeans.cluster_centers_
plt.scatter(centroids[:,0],centroids[:,1],s=200,c='black',marker='x',label='Centroids')
plt.title('Customer Segments (K-Means)')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.grid(True)
plt.show()
```
